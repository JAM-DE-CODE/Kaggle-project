```python
# Importing the pandas module
import pandas as pd
import matplotlib.pyplot as plt

# Reading in the sales data
df = pd.read_csv('data/sales_data.csv', parse_dates=['date'])

# Take a look at the first datapoints
df.head()
```

```python
df.head()
```

```python
df.describe()
```

```python
# Convert the total column to integer
df['total'] = df['total'].astype(int)
df
```

```python
# Question 1
# The total sales of each payment method.
total_sales = df.groupby('payment')[['total']].sum()
total_sales
```

```python
# Question 2
# The average unit price for each product line?

avg_unit = df.groupby('product_line')['unit_price'].mean()
avg_unit
```

```python
# Quest 3a
# Create plots to visualize findings for questions 1 and 2
# Plot a bar chart to visualize the finding for question 1

total_sales.plot(kind='bar')
plt.xlabel("Payment method")
plt.ylabel("Total Sales")
plt.title("Total sales of each payment method")
plt.show()
```

```python
# Question 3b
# Plot a bar chart to visualize the finding for question 2

avg_unit.plot(kind='bar')
plt.xlabel("Product line")
plt.ylabel("Average price")
plt.title("The average unit price for each product line")
plt.show()
```

```python
# Question 4a
# Investigate further (e.g., average purchase value by client type, total purchase value by product line, etc.)

avg_purchase = df.groupby('client_type')['unit_price'].mean()
avg_purchase
```

```python
# Plot a bar chart to visualize the finding for question 4a

avg_purchase.plot(kind='bar')
plt.xlabel("Client type")
plt.ylabel("Average purchase value")
plt.title("The average purchase value by client type")
plt.show()
```

```python
# Question 4b
# The total total purchase value by product line.
total_purchase = df.groupby('product_line')[['total']].sum()
total_purchase
```

```python
# Plot a bar chart to visualize the finding for question 4b

total_purchase.plot(kind='bar')
plt.xlabel("Product line")
plt.ylabel("Total")
plt.title("The total purchase value by product line")
plt.show()
```
